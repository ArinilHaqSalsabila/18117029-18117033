```python
import pandas as pd
```

```python
# Mengimport data peserta SNMPTN dalam Excel 
data = pd.read_excel('bigdata.xlsx',index_col = 0)
data
```

```python
# Ambil Data Peserta SNMPTN yang memilih Fakultas STEI, Kode = 3321073
stei = data[data.Pilihan == 3321073]
stei
```

```python
# Membuat Variabel Baru Data Peserta SNMPTN yang memilih STEI dengan menghilangkan Kolom 'Pilihan'
nilai_STEI = stei
del nilai_STEI['Pilihan']
nilai_STEI
```

```python
# Rata-rata Nilai Rapor Semua Semester
rata_STEI = nilai_STEI.mean(axis = 1)
rata_STEI = pd.DataFrame(rata_STEI, columns = ['Rata'])
rata_STEI
```

```python
# Merge Variabel Nilai STEI dengan Rata-rata yang Telah Dicari Sebelumnya
nilai_STEI = nilai_STEI.merge(rata_STEI, how = 'right', right_on = 'No. Peserta', left_on = 'No. Peserta')
nilai_STEI
```

```python
# Filter 1 : Nilai Rata-rata harus >= 80

filter1_STEI = nilai_STEI[nilai_STEI.Rata >= 80]
filter1_STEI
```

```python
#Filter 2 : Rata-rata Fisika 1-5 dan Matematika 1-5 lebih dari sama dengan 85

#Rata-rata Matematika
rataMAT_STEI = filter1_STEI[['Nama Siswa','MAT1','MAT2','MAT3','MAT4','MAT5']]
rataMAT_STEI = rataMAT_STEI.mean(axis = 1)
rataMAT_STEI = pd.DataFrame(rataMAT_STEI, columns = ['RataMAT'])

#Rata-rata Fisika
rataFIS_STEI = filter1_STEI[['Nama Siswa','FIS1','FIS2','FIS3','FIS4','FIS5']]
rataFIS_STEI = rataFIS_STEI.mean(axis = 1)
rataFIS_STEI = pd.DataFrame(rataFIS_STEI, columns = ['RataFIS'])

#Merge Filter 1 dengan rata-rata Matematika
filter2_STEI = filter1_STEI.merge(rataMAT_STEI, how = 'right', right_on = 'No. Peserta', left_on = 'No. Peserta')
#Merge lagi dengan rata-rata Fisika
filter2_STEI = filter2_STEI.merge(rataFIS_STEI,how = 'right', right_on = 'No. Peserta', left_on = 'No. Peserta')

#Melakukan Filter 2
filter2_STEI = filter2_STEI[filter2_STEI.RataFIS >= 85]
filter2_STEI = filter2_STEI[filter2_STEI.RataMAT >= 85]
filter2_STEI
```

```python
#Filter 3 : Nilai Matematika dan Fisika harus selalu minimal turun 1 nilai dari semester sebelumnya

filter3_STEI = filter2_STEI[(filter2_STEI.MAT5 >= (filter2_STEI.MAT4-1)) & (filter2_STEI.MAT4 >= (filter2_STEI.MAT3-1)) & (filter2_STEI.MAT3 >= (filter2_STEI.MAT2-1)) & (filter2_STEI.MAT2 >= (filter2_STEI.MAT1-1)) ]
filter3_STEI = filter2_STEI[(filter2_STEI.FIS5 >= (filter2_STEI.FIS4-1)) & (filter2_STEI.FIS4 >= (filter2_STEI.FIS3-1)) & (filter2_STEI.FIS3 >= (filter2_STEI.FIS2-1)) & (filter2_STEI.FIS2 >= (filter2_STEI.FIS1-1)) ]
filter3_STEI
```

```python
#Filter 4 : Meloloskan prestasi Internasional bidang Sains
lolos_Inter = filter3_STEI[(filter3_STEI.TingkatPrestasi == 'Internasional') & (filter3_STEI.BidangPrestasi == 'Sains')]

#Sisa peserta dari filter 4
filter4_STEI = filter3_STEI[(filter3_STEI.TingkatPrestasi != 'Internasional') | (filter3_STEI.BidangPrestasi != 'Sains')]
filter4_STEI
```

```python
#Filter 5 : Memberi nilai tambahan => Nasional = 2 poin; Provinsi = 1 poin; Kota/Kabupaten = 0,5
poin_STEI = (filter4_STEI.RataMAT + filter4_STEI.RataFIS)/2
poin_STEI = pd.DataFrame(poin_STEI, columns = ['Poin'])

filter5_STEI = filter4_STEI.merge(poin_STEI, how = 'right', right_on = 'No. Peserta', left_on = 'No. Peserta')
filter5_STEI
```

```python
# Menghitung nilai akhir, yaitu Rata-rata Matematika dan Fisika ditambah poin prestasi

filter5_STEI.loc[(filter5_STEI['TingkatPrestasi'] == 'Nasional') & (filter5_STEI['BidangPrestasi'] == 'Sains'), 'NilaiAkhir'] = filter5_STEI.Poin + 2 
filter5_STEI.loc[(filter5_STEI['TingkatPrestasi'] == 'Provinsi') & (filter5_STEI['BidangPrestasi'] == 'Sains'), 'NilaiAkhir'] = filter5_STEI.Poin + 1
filter5_STEI.loc[(filter5_STEI['TingkatPrestasi'] == 'Kota/Kabupaten') & (filter5_STEI['BidangPrestasi'] == 'Sains'), 'NilaiAkhir'] = filter5_STEI.Poin + 0.5
filter5_STEI.loc[filter5_STEI['BidangPrestasi'] != 'Sains', 'NilaiAkhir'] = filter5_STEI.Poin
filter5_STEI
```

```python
#Mengurutkan berdasarkan Nilai Akhir

filter5_STEI = filter5_STEI.sort_values(by='NilaiAkhir', ascending=False)
filter5_STEI
```

```python
# Data Peserta SNMPTN yang lolos STEI karena memiliki Prestasi Tingkat Internasional bidang Sains

lolos_Inter

# Didapat ada 3 orang yang lolos
```

```python
#Peserta yang lolos SNMPTN STEI ITB sebanyak 285 peserta
kuota_STEI = 285

# Mengurangi variabel Kuota STEI dengan jumlah yang lolos karena prestasi Sains Internasional
kuota_STEI = kuota_STEI - len(lolos_Inter)
```

```python
# Meloloskan Pemilih STEI dengan Ranking teratas sesuai sisa kuota lolos

filter5_STEI = filter5_STEI[:(kuota_STEI)]
filter5_STEI
```

```python
# Menggabungkan data yang lolos dengan yang lolos karena Prestasi Sains tingkat Internasional

lolos_STEI = filter5_STEI.append(lolos_Inter , ignore_index=True, sort=False)
del lolos_STEI['NilaiAkhir']
del lolos_STEI['Poin']
del lolos_STEI['RataFIS']
del lolos_STEI['RataMAT']
lolos_STEI
```
